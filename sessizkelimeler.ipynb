```python
# Gerekli kütüphanelerin içe aktarılması
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, backend as K
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
import numpy as np
import os
import matplotlib.pyplot as plt

# GPU kontrolü - Kaggle otomatik yönetir
physical_devices = tf.config.list_physical_devices('GPU')
print(f"Kullanılabilir GPU sayısı: {len(physical_devices)}")
if len(physical_devices) > 0:
    print("GPU aktif")
else:
    print("CPU kullanılacak")

# Squash fonksiyonu
def squash(vectors, axis=-1):
    s_squared_norm = K.sum(K.square(vectors), axis, keepdims=True)
    scale = s_squared_norm / (1 + s_squared_norm)
    return scale * vectors / K.sqrt(s_squared_norm + K.epsilon())

# Primary Capsule katmanı
class PrimaryCapsule(layers.Layer):
    def __init__(self, dim_capsule, n_channels, kernel_size, strides, padding, **kwargs):
        super(PrimaryCapsule, self).__init__(**kwargs)
        self.dim_capsule = dim_capsule
        self.n_channels = n_channels
        self.kernel_size = kernel_size
        self.strides = strides
        self.padding = padding
        self.output_num_capsule = None

    def build(self, input_shape):
        self.conv = layers.Conv2D(
            filters=self.dim_capsule * self.n_channels,
            kernel_size=self.kernel_size,
            strides=self.strides,
            padding=self.padding,
            name='primary_conv'
        )
        # Conv layer'ı build et
        self.conv.build(input_shape)
        
        # Conv çıktı boyutunu hesapla
        conv_output_shape = self.conv.compute_output_shape(input_shape)
        # conv_output_shape: (batch, height, width, channels)
        
        # Capsule sayısını hesapla
        if conv_output_shape[1] is not None and conv_output_shape[2] is not None:
            self.output_num_capsule = conv_output_shape[1] * conv_output_shape[2] * self.n_channels
        
        super(PrimaryCapsule, self).build(input_shape)
    
    def compute_output_shape(self, input_shape):
        # Conv çıktı boyutunu hesapla
        conv_output_shape = self.conv.compute_output_shape(input_shape)
        
        # Capsule formatına dönüştür
        if conv_output_shape[1] is not None and conv_output_shape[2] is not None:
            num_capsule = conv_output_shape[1] * conv_output_shape[2] * self.n_channels
            return (input_shape[0], num_capsule, self.dim_capsule)
        else:
            return (input_shape[0], None, self.dim_capsule)

    def call(self, inputs):
        outputs = self.conv(inputs)
        # outputs shape: [batch, height, width, channels]
        
        batch_size = tf.shape(outputs)[0]
        # Yeniden şekillendirme: [batch, height*width*n_channels, dim_capsule]
        outputs = K.reshape(outputs, [batch_size, -1, self.dim_capsule])
        
        return squash(outputs)
    
    def get_config(self):
        config = {
            'dim_capsule': self.dim_capsule,
            'n_channels': self.n_channels,
            'kernel_size': self.kernel_size,
            'strides': self.strides,
            'padding': self.padding
        }
        base_config = super(PrimaryCapsule, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

# Kapsül katmanı
class CapsuleLayer(layers.Layer):
    def __init__(self, num_capsule, dim_capsule, routings=3, **kwargs):
        super(CapsuleLayer, self).__init__(**kwargs)
        self.num_capsule = num_capsule
        self.dim_capsule = dim_capsule
        self.routings = routings
        self.input_num_capsule = None
        self.input_dim_capsule = None

    def build(self, input_shape):
        # Shape kontrolü ve çıkarımı
        if input_shape[1] is not None:
            self.input_num_capsule = int(input_shape[1])
        else:
            raise ValueError(f"Input shape[1] cannot be None. Got input_shape: {input_shape}")
            
        if input_shape[2] is not None:
            self.input_dim_capsule = int(input_shape[2])
        else:
            raise ValueError(f"Input shape[2] cannot be None. Got input_shape: {input_shape}")
        
        # Ağırlık matrisi
        self.W = self.add_weight(
            shape=[self.num_capsule, self.input_num_capsule, self.dim_capsule, self.input_dim_capsule],
            initializer='glorot_uniform',
            trainable=True,
            name='W'
        )
            
        super(CapsuleLayer, self).build(input_shape)
    
    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.num_capsule, self.dim_capsule)

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        
        # inputs shape: [batch_size, input_num_capsule, input_dim_capsule]
        inputs_expand = K.expand_dims(inputs, axis=1)
        # shape: [batch_size, 1, input_num_capsule, input_dim_capsule]
        
        inputs_tile = K.tile(inputs_expand, [1, self.num_capsule, 1, 1])
        # shape: [batch_size, num_capsule, input_num_capsule, input_dim_capsule]
        
        # W shape: [num_capsule, input_num_capsule, dim_capsule, input_dim_capsule]
        # inputs_tile: [batch_size, num_capsule, input_num_capsule, input_dim_capsule]
        # Reshape inputs_tile için: [batch, num_capsule, input_num_capsule, 1, input_dim_capsule]
        inputs_tile_expanded = K.expand_dims(inputs_tile, axis=3)
        
        # W'yi tile et: [1, num_capsule, input_num_capsule, dim_capsule, input_dim_capsule]
        W_tiled = K.expand_dims(self.W, axis=0)
        
        # Matris çarpımı: son boyut üzerinden
        # [batch, num_capsule, input_num_capsule, 1, input_dim_capsule] x 
        # [1, num_capsule, input_num_capsule, dim_capsule, input_dim_capsule]
        inputs_hat = tf.reduce_sum(inputs_tile_expanded * W_tiled, axis=-1)
        # shape: [batch_size, num_capsule, input_num_capsule, dim_capsule]
        
        # Routing logitleri başlangıç değerleri
        b = tf.zeros([batch_size, self.num_capsule, self.input_num_capsule])
        
        # Dinamik routing
        for i in range(self.routings):
            c = tf.nn.softmax(b, axis=1)
            # shape: [batch_size, num_capsule, input_num_capsule]
            
            c_expand = K.expand_dims(c, axis=-1)
            # shape: [batch_size, num_capsule, input_num_capsule, 1]
            
            # Ağırlıklı toplam
            s = tf.reduce_sum(c_expand * inputs_hat, axis=2)
            # shape: [batch_size, num_capsule, dim_capsule]
            
            # Squash aktivasyonu
            outputs = squash(s, axis=-1)
            # shape: [batch_size, num_capsule, dim_capsule]
            
            if i < self.routings - 1:
                # Routing logitlerini güncelle
                outputs_expand = K.expand_dims(outputs, axis=2)
                # shape: [batch_size, num_capsule, 1, dim_capsule]
                
                # Agreement hesaplama
                agreement = tf.reduce_sum(inputs_hat * outputs_expand, axis=-1)
                # shape: [batch_size, num_capsule, input_num_capsule]
                
                b = b + agreement
        
        return outputs
    
    def get_config(self):
        config = {
            'num_capsule': self.num_capsule,
            'dim_capsule': self.dim_capsule,
            'routings': self.routings
        }
        base_config = super(CapsuleLayer, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

# Model oluşturma
def create_model(learning_rate=0.001, num_classes=10):
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    
    for layer in base_model.layers[:-20]:
        layer.trainable = False
    
    x = base_model.output
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    primary_caps = PrimaryCapsule(
        dim_capsule=8,
        n_channels=16,
        kernel_size=3,
        strides=2,
        padding='valid'
    )(x)
    
    digit_caps = CapsuleLayer(
        num_capsule=num_classes,
        dim_capsule=16,
        routings=3
    )(primary_caps)
    
    out_caps = layers.Lambda(lambda z: K.sqrt(K.sum(K.square(z), axis=-1)))(digit_caps)
    
    model = models.Model(inputs=base_model.input, outputs=out_caps)
    model.compile(
        optimizer=optimizers.Adam(learning_rate=learning_rate),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    return model

# Veri hazırlama
def prepare_data(data_dir, batch_size=32):
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        rotation_range=10,
        width_shift_range=0.1,
        height_shift_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True
    )

    train_generator = train_datagen.flow_from_directory(
        data_dir,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        subset='training',
        shuffle=True
    )

    validation_generator = train_datagen.flow_from_directory(
        data_dir,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation',
        shuffle=True
    )

    num_classes = train_generator.num_classes
    print(f"\nToplam sınıf sayısı: {num_classes}")
    print(f"Sınıf isimleri: {list(train_generator.class_indices.keys())}")
    print(f"Eğitim örnek sayısı: {train_generator.samples}")
    print(f"Doğrulama örnek sayısı: {validation_generator.samples}")

    return train_generator, validation_generator, num_classes

# Eğitim fonksiyonu
def train_model(data_dir, epochs=20, learning_rate=0.0001, batch_size=32):
    # Veri hazırlama
    train_gen, val_gen, num_classes = prepare_data(data_dir, batch_size)
    
    # Model oluşturma
    print("\nModel oluşturuluyor...")
    model = create_model(learning_rate=learning_rate, num_classes=num_classes)
    print(model.summary())
    
    # Callbacks
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True,
            verbose=1
        ),
        ModelCheckpoint(
            '/kaggle/working/best_model.h5',
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-7,
            verbose=1
        )
    ]
    
    # Model eğitimi
    print("\nModel eğitimi başlıyor...")
    history = model.fit(
        train_gen,
        validation_data=val_gen,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )
    
    # Sonuçları görselleştirme
    plot_training_history(history)
    
    return model, history

# Eğitim geçmişini görselleştirme
def plot_training_history(history):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # Accuracy grafiği
    ax1.plot(history.history['accuracy'], label='Eğitim')
    ax1.plot(history.history['val_accuracy'], label='Doğrulama')
    ax1.set_title('Model Doğruluğu')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Doğruluk')
    ax1.legend()
    ax1.grid(True)
    
    # Loss grafiği
    ax2.plot(history.history['loss'], label='Eğitim')
    ax2.plot(history.history['val_loss'], label='Doğrulama')
    ax2.set_title('Model Kaybı')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Kayıp')
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\nEn iyi doğrulama doğruluğu:", max(history.history['val_accuracy']))

# Test verisi ile tahmin (Opsiyonel)
def predict_on_test(model_path, test_dir):
    """
    Test verisi üzerinde tahmin yapar
    """
    # Custom objects tanımla
    custom_objects = {
        'PrimaryCapsule': PrimaryCapsule,
        'CapsuleLayer': CapsuleLayer,
        'squash': squash
    }
    
    # Modeli yükle
    model = tf.keras.models.load_model(model_path, custom_objects=custom_objects)
    
    # Test verisi için generator
    test_datagen = ImageDataGenerator(rescale=1./255)
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
    
    # Tahmin
    predictions = model.predict(test_generator)
    
    # Sonuçları döndür
    return predictions, test_generator

# Ana fonksiyon
def main():
    # KAGGLE VERİ YOLU
    data_dir = '/kaggle/input/tr-sign-language/tr_signLanguage_dataset/train/'
    
    # Veri yolunu kontrol et
    if not os.path.exists(data_dir):
        print(f"HATA: {data_dir} bulunamadı!")
        print("Lütfen veri setini doğru şekilde ekleyin.")
        return
    
    print(f"Veri klasörü bulundu: {data_dir}")
    print(f"Sınıflar: {os.listdir(data_dir)}")
    
    # Modeli eğit
    model, history = train_model(
        data_dir=data_dir,
        epochs=20,
        learning_rate=0.0001,
        batch_size=32
    )
    
    print("\nEğitim tamamlandı!")
    print("Model /kaggle/working/best_model.h5 konumuna kaydedildi.")
    print("Modeli indirmek için: Files -> Output -> best_model.h5")

if __name__ == "__main__":
    main()
```

```python
# ============================================================================
# TEST KODU - Eğitilmiş Modeli Test Etmek İçin
# ============================================================================

import tensorflow as tf
from tensorflow.keras import layers, backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

print("Kütüphaneler yüklendi\n")

# ============================================================================
# CUSTOM LAYER TANIMLARI
# ============================================================================

def squash(vectors, axis=-1):
    s_squared_norm = K.sum(K.square(vectors), axis, keepdims=True)
    scale = s_squared_norm / (1 + s_squared_norm)
    return scale * vectors / K.sqrt(s_squared_norm + K.epsilon())

class PrimaryCapsule(layers.Layer):
    def __init__(self, dim_capsule, n_channels, kernel_size, strides, padding, **kwargs):
        super(PrimaryCapsule, self).__init__(**kwargs)
        self.dim_capsule = dim_capsule
        self.n_channels = n_channels
        self.kernel_size = kernel_size
        self.strides = strides
        self.padding = padding
        self.output_num_capsule = None

    def build(self, input_shape):
        self.conv = layers.Conv2D(
            filters=self.dim_capsule * self.n_channels,
            kernel_size=self.kernel_size,
            strides=self.strides,
            padding=self.padding,
            name='primary_conv'
        )
        self.conv.build(input_shape)
        conv_output_shape = self.conv.compute_output_shape(input_shape)
        if conv_output_shape[1] is not None and conv_output_shape[2] is not None:
            self.output_num_capsule = conv_output_shape[1] * conv_output_shape[2] * self.n_channels
        super(PrimaryCapsule, self).build(input_shape)
    
    def compute_output_shape(self, input_shape):
        conv_output_shape = self.conv.compute_output_shape(input_shape)
        if conv_output_shape[1] is not None and conv_output_shape[2] is not None:
            num_capsule = conv_output_shape[1] * conv_output_shape[2] * self.n_channels
            return (input_shape[0], num_capsule, self.dim_capsule)
        else:
            return (input_shape[0], None, self.dim_capsule)

    def call(self, inputs):
        outputs = self.conv(inputs)
        batch_size = tf.shape(outputs)[0]
        outputs = K.reshape(outputs, [batch_size, -1, self.dim_capsule])
        return squash(outputs)
    
    def get_config(self):
        config = {
            'dim_capsule': self.dim_capsule,
            'n_channels': self.n_channels,
            'kernel_size': self.kernel_size,
            'strides': self.strides,
            'padding': self.padding
        }
        base_config = super(PrimaryCapsule, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

class CapsuleLayer(layers.Layer):
    def __init__(self, num_capsule, dim_capsule, routings=3, **kwargs):
        super(CapsuleLayer, self).__init__(**kwargs)
        self.num_capsule = num_capsule
        self.dim_capsule = dim_capsule
        self.routings = routings
        self.input_num_capsule = None
        self.input_dim_capsule = None

    def build(self, input_shape):
        if input_shape[1] is not None:
            self.input_num_capsule = int(input_shape[1])
        else:
            raise ValueError(f"Input shape[1] cannot be None. Got input_shape: {input_shape}")
            
        if input_shape[2] is not None:
            self.input_dim_capsule = int(input_shape[2])
        else:
            raise ValueError(f"Input shape[2] cannot be None. Got input_shape: {input_shape}")
        
        self.W = self.add_weight(
            shape=[self.num_capsule, self.input_num_capsule, self.dim_capsule, self.input_dim_capsule],
            initializer='glorot_uniform',
            trainable=True,
            name='W'
        )
        super(CapsuleLayer, self).build(input_shape)
    
    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.num_capsule, self.dim_capsule)

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        inputs_expand = K.expand_dims(inputs, axis=1)
        inputs_tile = K.tile(inputs_expand, [1, self.num_capsule, 1, 1])
        inputs_tile_expanded = K.expand_dims(inputs_tile, axis=3)
        W_tiled = K.expand_dims(self.W, axis=0)
        inputs_hat = tf.reduce_sum(inputs_tile_expanded * W_tiled, axis=-1)
        b = tf.zeros([batch_size, self.num_capsule, self.input_num_capsule])
        
        for i in range(self.routings):
            c = tf.nn.softmax(b, axis=1)
            c_expand = K.expand_dims(c, axis=-1)
            s = tf.reduce_sum(c_expand * inputs_hat, axis=2)
            outputs = squash(s, axis=-1)
            
            if i < self.routings - 1:
                outputs_expand = K.expand_dims(outputs, axis=2)
                agreement = tf.reduce_sum(inputs_hat * outputs_expand, axis=-1)
                b = b + agreement
        
        return outputs
    
    def get_config(self):
        config = {
            'num_capsule': self.num_capsule,
            'dim_capsule': self.dim_capsule,
            'routings': self.routings
        }
        base_config = super(CapsuleLayer, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

print("Custom layer'lar tanımlandı\n")

# ============================================================================
# MODEL YÜKLEME VE TEST FONKSİYONU
# ============================================================================

def load_model_weights_only(model_path, test_dir):
    """Sadece ağırlıkları yükleyerek modeli test eder"""
    from tensorflow.keras.applications import MobileNetV2
    from tensorflow.keras import models, optimizers
    
    print("=" * 60)
    print("MODEL OLUŞTURULUYOR VE AĞIRLIKLAR YÜKLENIYOR")
    print("=" * 60)
    
    # Test datası için sınıf sayısını al
    temp_datagen = ImageDataGenerator(rescale=1./255)
    temp_generator = temp_datagen.flow_from_directory(
        test_dir,
        target_size=(224, 224),
        batch_size=1,
        class_mode='categorical',
        shuffle=False
    )
    num_classes = temp_generator.num_classes
    print(f"Sınıf sayısı: {num_classes}")
    
    # Modeli yeniden oluştur
    base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    
    for layer in base_model.layers[:-20]:
        layer.trainable = False
    
    x = base_model.output
    x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    
    primary_caps = PrimaryCapsule(
        dim_capsule=8,
        n_channels=16,
        kernel_size=3,
        strides=2,
        padding='valid'
    )(x)
    
    digit_caps = CapsuleLayer(
        num_capsule=num_classes,
        dim_capsule=16,
        routings=3
    )(primary_caps)
    
    out_caps = layers.Lambda(
        lambda z: K.sqrt(K.sum(K.square(z), axis=-1)),
        output_shape=(num_classes,)
    )(digit_caps)
    
    model = models.Model(inputs=base_model.input, outputs=out_caps)
    model.compile(
        optimizer=optimizers.Adam(learning_rate=0.0001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    # Ağırlıkları yükle
    print(f"\nAğırlıklar yükleniyor: {model_path}")
    model.load_weights(model_path)
    print("Ağırlıklar başarıyla yüklendi\n")
    
    return model, temp_generator

def evaluate_model(model, test_generator):
    """Modeli test eder ve sonuçları gösterir"""
    
    print("=" * 60)
    print("TEST VERİSİ BİLGİLERİ")
    print("=" * 60)
    print(f"Test örnek sayısı: {test_generator.samples}")
    print(f"Test sınıf sayısı: {test_generator.num_classes}")
    print(f"Sınıflar: {list(test_generator.class_indices.keys())}\n")
    
    # Model değerlendirmesi
    print("=" * 60)
    print("MODEL DEĞERLENDİRMESİ")
    print("=" * 60)
    test_loss, test_accuracy = model.evaluate(test_generator, verbose=1)
    
    print(f"\n{'='*60}")
    print("TEST SONUÇLARI")
    print(f"{'='*60}")
    print(f"Test Kaybı (Loss):      {test_loss:.4f}")
    print(f"Test Doğruluğu (Acc):   {test_accuracy*100:.2f}%")
    print(f"{'='*60}\n")
    
    # Detaylı tahminler
    print("=" * 60)
    print("DETAYLI TAHMİNLER")
    print("=" * 60)
    predictions = model.predict(test_generator, verbose=1)
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = test_generator.classes
    
    correct = np.sum(predicted_classes == true_classes)
    incorrect = np.sum(predicted_classes != true_classes)
    
    print(f"\nDoğru tahmin:  {correct}/{len(true_classes)} ({100*correct/len(true_classes):.2f}%)")
    print(f"Yanlış tahmin: {incorrect}/{len(true_classes)} ({100*incorrect/len(true_classes):.2f}%)\n")
    
    # Confusion Matrix
    cm = confusion_matrix(true_classes, predicted_classes)
    
    print("=" * 60)
    print("CONFUSION MATRIX OLUŞTURULUYOR")
    print("=" * 60)
    
    plt.figure(figsize=(16, 14))
    class_names = list(test_generator.class_indices.keys())
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, 
                yticklabels=class_names,
                cbar_kws={'label': 'Örnek Sayısı'},
                linewidths=0.5)
    plt.title(f'Test Dataseti - Confusion Matrix\n(Test Doğruluğu: {test_accuracy*100:.2f}%)', 
              fontsize=18, pad=20, fontweight='bold')
    plt.ylabel('Gerçek Sınıf', fontsize=14, fontweight='bold')
    plt.xlabel('Tahmin Edilen Sınıf', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig('/kaggle/working/test_confusion_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()
    print("Confusion matrix kaydedildi: /kaggle/working/test_confusion_matrix.png\n")
    
    # Classification Report
    print("=" * 60)
    print("SINIF BAZINDA PERFORMANS RAPORU")
    print("=" * 60)
    report = classification_report(true_classes, predicted_classes, 
                                   target_names=class_names, digits=4)
    print(report)
    
    # En çok karıştırılan sınıflar
    print("\n" + "=" * 60)
    print("EN ÇOK KARIŞTIRILAN SINIFLAR (Top 10)")
    print("=" * 60)
    
    confusion_pairs = []
    for i in range(len(class_names)):
        for j in range(len(class_names)):
            if i != j and cm[i][j] > 0:
                confusion_pairs.append((class_names[i], class_names[j], cm[i][j]))
    
    confusion_pairs.sort(key=lambda x: x[2], reverse=True)
    
    if confusion_pairs:
        for idx, (true_class, pred_class, count) in enumerate(confusion_pairs[:10], 1):
            print(f"{idx:2d}. '{true_class}' -> '{pred_class}' olarak yanlış tahmin: {count} kez")
    else:
        print("Hiç yanlış tahmin yok! Mükemmel performans!")
    
    print("\n" + "=" * 60)
    
    return test_accuracy, predictions, cm

# ============================================================================
# TEST ÇALIŞTIRMA
# ============================================================================

test_dir = '/kaggle/input/tr-sign-language/tr_signLanguage_dataset/test/'
model_path = '/kaggle/input/sessizkelimelermodel/best_model.h5'

print("TEST DEĞERLENDİRMESİ BAŞLIYOR\n")
print("=" * 60)

# Modeli yükle
model, test_generator = load_model_weights_only(model_path, test_dir)

# Test et
test_accuracy, predictions, cm = evaluate_model(model, test_generator)

print("\n" + "=" * 60)
print("TEST DEĞERLENDİRMESİ TAMAMLANDI")
print("=" * 60)
print(f"Final Test Doğruluğu: {test_accuracy*100:.2f}%")
print(f"Confusion matrix: /kaggle/working/test_confusion_matrix.png")
print("=" * 60)
```
